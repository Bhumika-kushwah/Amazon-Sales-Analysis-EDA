```python

cat_cols=[col for col in df.columns if df[col].dtype =="object"]
num_cols=[col for col in df.columns if df[col].dtype !="object"]
```

```python
#for col in cat_cols:
#    print(f"{col} has {df[col].unique()} values \n")
for col in num_cols:
    print(f"{col} has {df[col].unique()}values \n")
```

```python
#univariate anlaysis
for num_column in num_cols:
    plt.figure(figsize=(8,5))
    sns.histplot(df[num_column],kde=True,color='skyblue')
    plt.title(f"univariate analysis - {num_column}")
    
```

```python
#avrage rating for each product category
df.groupby('category')['rating'].mean()
```

```python
top10=df.groupby('category')['rating'].mean().sort_values(ascending=False).head(10).reset_index()
top10
plt.figure(figsize=(12,6))
sns.barplot(x='category',y='rating',data=top10)
plt.title('Top 10 Category by Avrage rating')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.xticks(rotation=75)
plt.show()
```

```python
#top rating count products by category
df.sort_values('rating_count',ascending=False).groupby('category').head(1)
```

```python
#discount vs actuall price
 

plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sns.histplot(df['actual_price'],label='Actual',kde=True)
plt.title('Acutal price')


plt.subplot(2,2,2)
sns.histplot(df['discounted_price'], label='Discounted', kde=True)
plt.title('Discount price')
plt.show()
```

```python
# What are the most popular product reviews
df.groupby('product_name')['review_title'].value_counts().head(5)
```

```python
# What is the correlation between discounted_price and rating
corr=df[['discounted_price', 'rating']].corr()
sns.heatmap(corr,annot=True)
plt.show()
```
